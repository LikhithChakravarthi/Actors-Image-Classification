# Importing required Packages

```python
import tensorflow as tf
import os
import cv2
import imghdr
from matplotlib import pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
from tensorflow.keras.metrics import Precision,Recall,BinaryAccuracy
from tensorflow.keras.models import load_model
```

# Loading and Cleaning Data

```python
data_dir='data'
```

```python
image_exts=['jpeg','jpg','bmp','png']
```

```python
classes=os.listdir(data_dir)
```

```python
classes
```

Code written below will remove any inappropriate images if present

```python
for image_class in os.listdir(data_dir): 
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            os.remove(image_path)
```

Here we are creating a dataset out of the 'data' directory. This is the beat bart of keras, we can directly creating dataset out of the files present in the directory. It also maps the class labels.

```python
data=tf.keras.utils.image_dataset_from_directory("data")
```

```python
data
```

```python
data_iterator=data.as_numpy_iterator()
```

```python
batch=data_iterator.next()
```

We can fetch the data batchwise using the above code

```python
fig,ax=plt.subplots(ncols=4,figsize=(20,20))
for idx,img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

I the below code we are scaling the data for convenience

```python
data=data.map(lambda x,y:(x/255,y))
```

```python
data.as_numpy_iterator().next()[0].shape
```

# Splitting The Data

```python
train_size=int(len(data)*0.7)
val_size=int(len(data)*0.2)
test_size=int(len(data)*0.1)
```

```python
train=data.take(train_size)
val=data.skip(train_size).take(round(val_size))
test=data.skip(train_size+val_size).take(round(test_size))
```

```python
train
```

# Creating Model

```python
model=Sequential()
```

```python
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss="sparse_categorical_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdirs="logs"
```

# Training Model

```python
tensorboard_callbacks=tf.keras.callbacks.TensorBoard(log_dir=logdirs)
```

```python
hist=model.fit(train,epochs=12,validation_data=val,callbacks=[tensorboard_callbacks])
```

```python
fig=plt.figure()
plt.plot(hist.history['loss'],color='red',label="Loss")
plt.plot(hist.history['val_loss'],color='blue',label="Val_Loss")
plt.suptitle('Loss',fontsize=20)
plt.legend()
plt.show()
```

Im the above graph we can see that the loss is constantly decreasing, which shows the model is working well. If the loss is decreasing and the Val_Loss is not, then it is an indication that the model is overfitting. In that case we have to change the configuration of model.

```python
fig=plt.figure()
plt.plot(hist.history['accuracy'],color='red',label="Accuracy")
plt.plot(hist.history['val_accuracy'],color='blue',label="Val_Accuracy")
plt.suptitle("Accuracy",fontsize=20)
plt.legend()
plt.show()
```

```python
model.evaluate(test)
```

The Accuracy of the model is 97% which is pretty good. If we want to increase it more, then we have to increase the training data, or we can increase the epochs value, or change the configuration of the model.
